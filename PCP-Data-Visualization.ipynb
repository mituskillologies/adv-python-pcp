#### Bar plot

```python
studs= [57,67,56,78,45]
course= ['C','C++','Python','R','Ruby']
```

```python
plt.title('Total Enrollment')
plt.xlabel('Courses')
plt.ylabel('Students')
plt.grid()
plt.bar(course,studs, color = 'black')
```

#### pie

```python
plt.figure(figsize=(5,5))
plt.title('Students Enrolled')
plt.pie(studs, labels= course, autopct='%2.3f%%',
        shadow=True);
```

#### Histogram

```python
aa=[35,53,67,89,34,29,45,75,74,23,27,83]
```

```python
plt.title('Range of values')
plt.xticks(range(0,100,10))
plt.xlabel('range')
plt.ylabel('count')
plt.grid()
plt.hist(aa, bins=range(0,100,10),color = 'black')
```
